```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('dataset.csv')

# summary_date to ISO date format
df['summary_date'] = pd.to_datetime(df['summary_date'], format='%Y-%m-%d')

# group info by campaign_id
grouped = df.groupby('campaign_id').agg({'impressions': 'sum', 'clicks': 'sum', 'installs': 'sum', 'spend': 'sum'})


# CTR and CPI
grouped['CTR'] = grouped['clicks'] / grouped['impressions']
grouped['CPI'] = grouped['spend'] / grouped['installs']

# Display the report
grouped
```

```python
# CTR per Campaing
fig, ax = plt.subplots(figsize=(10, 8))
sns.barplot(x=grouped.index, y='CTR', data=grouped, color='green')
sns.despine()

ax.set_title('CTR per Campaing')
ax.set_xlabel('Campaing')
ax.set_ylabel('CTR')

plt.show()
```
